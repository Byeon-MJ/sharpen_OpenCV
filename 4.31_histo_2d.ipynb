```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 컬러 스타일을 1.x 스타일로 사용
plt.style.use('classic')
img = cv2.imread('./img/mountain.jpg')
```

```python
# 원본 이미지 확인
plt.imshow(img[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# 2D 히스토그램 출력
fig = plt.figure(figsize=(16, 8))
plt.subplot(131)
hist = cv2.calcHist([img], [0, 1], None, [32, 32], [0, 256, 0, 256])
p = plt.imshow(hist)
plt.title('Blue and Green')
plt.colorbar(p)

plt.subplot(132)
hist = cv2.calcHist([img], [1, 2], None, [32, 32], [0, 256, 0, 256])
p = plt.imshow(hist)
plt.title('Green and Red')
plt.colorbar(p)

plt.subplot(133)
hist = cv2.calcHist([img], [0, 2], None, [32, 32], [0, 256, 0, 256])
p = plt.imshow(hist)
plt.title('Blue and Red')
plt.colorbar(p)

plt.tight_layout()
plt.show()
```
