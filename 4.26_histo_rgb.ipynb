```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이미지 읽기 및 출력
img = cv2.imread('./img/mountain.jpg')

plt.imshow(img[:, :, ::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# 히스토그램 계산 및 그리기
channels = cv2.split(img)
colors = ('b', 'g', 'r')

for (ch, color) in zip(channels, colors):
    hist = cv2.calcHist([ch], [0], None, [256], [0, 256])
    plt.plot(hist, color = color)
plt.show()
```
