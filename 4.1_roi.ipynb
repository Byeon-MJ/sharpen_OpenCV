```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./img/sunset.jpg')
plt.imshow(img[:, :, ::-1])
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# roi 좌표
x = 320
y = 150
w = 50
h = 50
roi = img[y:y+h, x:x+w]

print(roi.shape)
```

```python
# roi에 사각형 그리기
cv2.rectangle(roi, (0, 0), (h-1, w-1), (0, 255, 0))

cv2.imshow('img', img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
