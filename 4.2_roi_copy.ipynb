```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./img/sunset.jpg')
plt.imshow(img[:, :, ::-1])
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# roi 
x = 320
y = 150
w = 50
h = 50
roi = img[y:y+h, x:x+w]  # roi 
img2 = roi.copy()        # roi 
```

```python
#   roi ( 2 )
img[y:y+h, x+w:x+w+w] = roi

# 2    
cv2.rectangle(img, (x, y), (x+w+w, y+h), (0, 255, 0))

#  
cv2.imshow('img', img)
cv2.imshow('roi', img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# matplotlib 
plt.imshow(img[:,:,::-1])
plt.xticks([]), plt.yticks([])
plt.show()

fig = plt.figure(figsize=(3, 3))
plt.imshow(img2[:,:,::-1])
plt.xticks([]), plt.yticks([])
plt.show()
```
